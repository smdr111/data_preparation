#1) Quyida berilgan datasetni yuklab olib, ustun nomlarini koʻring va ularni maʼno jihatdan tushunarli shaklga keltiring. `Unnamed: 0` ustunini tashlab yuboring!

```python
import pandas as pd
import numpy as np

url = "https://raw.githubusercontent.com/alishermutalov/praktikum-datasets/refs/heads/praktikum/restaurant_sales.csv"
df = pd.read_csv(url)
df.head()
```

#3) Har bir ustundagi boʻsh qiymatlarni aniqlang va ularni toʻldirish strategiyasini ishlab chiqing.

```python
#javobni shu yerda yozing
df.isnull().sum()
```

```python
df.Price.replace(-1,np.nan,inplace=True)
```

```python
grps=df.groupby('Category').Price.mean().round()
for idx in grps.index:
  df[df.Category==idx]=df[df.Category==idx].fillna({'Price':grps[idx]})
```

```python
df[df.Price.isnull()]
```

```python
df.dropna(subset=['Price'],inplace=True)
df.isnull().sum()
```

```python
df['Payment Method'].fillna(method='ffill',inplace=True)
```

```python
ctg=dict(zip(df.Price,df.Category))
df.Category=df.Category.fillna(df.Price.map(ctg))
```

```python
df.isnull().sum()
```

```python
item=dict(zip(df.Price,df.Item))
df.Item=df.Item.fillna(df.Price.map(item))
```

```python
df.dropna(subset=['Item'],inplace=True)
```

```python
df.isnull().sum()
```

```python
df.Quantity.fillna(method='ffill',inplace=True)
```

```python
df['Order Total'].fillna(df.Price*df.Quantity,inplace=True)
```

```python
df.isnull().sum()
```
